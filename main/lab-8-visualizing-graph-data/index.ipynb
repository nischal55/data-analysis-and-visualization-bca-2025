# Graph Data

```python
import networkx as nx
import matplotlib.pyplot as plt
```

Directed Graph

```python
DG = nx.DiGraph()
DG.add_edges_from([('A', 'B'), ('B', 'C'), ('C', 'D')])

nx.draw(DG, with_labels=True, node_color='skyblue', node_size=800, arrows=True)
plt.title("Directed Graph (Unweighted)")
plt.show()
```

Undirected  graph

```python
UG = nx.Graph()
UG.add_edges_from([('A', 'B'), ('B', 'C'), ('C', 'A')])

nx.draw(UG, with_labels=True, node_color='lightgreen', node_size=800)
plt.title("Undirected Graph (Unweighted)")
plt.show()
```

```python
WG = nx.Graph()
WG.add_weighted_edges_from([
    ('A', 'B', 4),
    ('B', 'C', 2),
    ('A', 'C', 5),
    ('C', 'D', 1)
])

pos = nx.spring_layout(WG)
edge_labels = nx.get_edge_attributes(WG, 'weight')

nx.draw(WG, pos, with_labels=True, node_color='lightcoral', node_size=800)
nx.draw_networkx_edge_labels(WG, pos, edge_labels=edge_labels)
plt.title("Weighted Undirected Graph")
plt.show()
```

Path Finding With Dijkstra

```python
G = nx.Graph()
G.add_weighted_edges_from([
    ('A', 'B', 4),
    ('B', 'C', 2),
    ('A', 'C', 5),
    ('C', 'D', 1)
])

source = 'A'
target = 'D'
path = nx.dijkstra_path(G, source, target)
length = nx.dijkstra_path_length(G, source, target)

print(f"Shortest path from {source} to {target}: {path}")
print(f"Total path weight: {length}")

pos = nx.spring_layout(G)
edge_labels = nx.get_edge_attributes(G, 'weight')

path_edges = list(zip(path, path[1:]))

nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=800)
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), width=1)
nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='red', width=3)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.title("Dijkstra's Shortest Path")
plt.show()
```
