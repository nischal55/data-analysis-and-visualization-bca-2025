## Flow Data

```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import autocorrelation_plot
from pandas.plotting import lag_plot
```

Line Plot

```python
date_rng = pd.date_range(start='2023-01-01', periods=12, freq='M')
data = pd.Series([5, 7, 6, 9, 12, 14, 13, 15, 17, 16, 18, 20], index=date_rng)

data.plot(figsize=(8, 4), marker='o')
plt.title("Time Series - Line Plot")
plt.xlabel("Date")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

**Analysis**

A line plot is the most common way to visualize time series data. It shows trends, patterns, and changes in values over time, making it easy to spot upward or downward trends, spikes, and seasonality.

---

```python
result = seasonal_decompose(data, model='additive', period=4)
result.plot()
plt.suptitle("Seasonal Decomposition", fontsize=16)
plt.show()
```

**Analysis**

Seasonal decomposition splits a time series into trend, seasonal, and residual components. This helps identify repeating patterns, long-term direction, and random fluctuations in the data.

---

 Bar Chart

```python
data.plot(kind='bar', figsize=(10, 4))
plt.title("Time Series - Bar Chart")
plt.xlabel("Date")
plt.ylabel("Value")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Analysis**

A bar chart shows individual values over time using bars. It emphasizes the magnitude of each observation and is useful for highlighting discrete comparisons between time points.


---

Heatmap

```python
df = data.to_frame(name='value')
df['Month'] = df.index.month
df['Year'] = df.index.year

# Pivot to create month-year matrix
pivot = df.pivot(index="Month", columns="Year", values="value")

# Heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Time Series Heatmap (Month-Year)")
plt.show()
```

**Analysis**

A time series heatmap visualizes data intensity over two time dimensions, typically year and month. It is useful for spotting seasonal patterns, anomalies, or trends that repeat at the same time each year.

---

Autocorrelation Plot

```python

plt.figure(figsize=(6, 4))

# Line
data.plot(figsize=(8, 4), marker='o')
plt.title("Time Series - Line Plot")
plt.xlabel("Date")
plt.ylabel("Value")
plt.grid(True)
plt.show()

# Auto Correlation
autocorrelation_plot(data)

plt.title("Autocorrelation Plot")
plt.show()
```

**Analysis**

An autocorrelation plot shows how current values in a time series relate to past values at different lags. Peaks at specific lags suggest periodicity or seasonality in the data.

---

Lag Plot

```python
plt.figure(figsize=(6, 4))
lag_plot(data)
plt.title("Lag Plot")
plt.show()
```

**Analysis**

A lag plot compares a time series against a lagged version of itself. A visible pattern (like linear or clustered points) indicates autocorrelation, while random scatter suggests randomness.
