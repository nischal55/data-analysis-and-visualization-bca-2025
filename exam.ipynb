```python
import numpy as np
study_hours = np.array([10, 12, 20, 43, 57, 42, 23, 17, 68])
marks = np.array([42.2, 49.6, 58.5, 62.24, 78.71, 64.52, 60.23, 51.15, 92.24])

# Step 1: Identify independent (X) and dependent (Y) variables
X = study_hours  # Independent variable
Y = marks         # Dependent variable

# Step 2: Calculate means
mean_X = np.mean(X)
mean_Y = np.mean(Y)


print("mean_X", mean_X)
print("mean_Y", mean_Y)

xxb = X - mean_X
yyb = Y - mean_Y
num = np.sum(xxb*yyb)
den = np.sum((X - mean_X) ** 2)


# print("xMinusXBar =", xxb)
# print("yMinusYBar =", yyb)
print("E(xxb*yyb) =", np.sum(xxb*yyb))
print("E(xi-xbar)**2 =", np.sum((X - mean_X) ** 2))
print("m =", num/den)

```

```python
# Re-attempt the animation with corrected point update using sequences for x and y

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import PillowWriter


# Define the function and its derivative
def f(x):
    return x**2

def f_prime(x):
    return 2*x

# Gradient descent parameters
eta = 0.1  # learning rate
x0 = 4     # starting point
iterations = 25

# Perform gradient descent and store path
x_vals = [x0]
y_vals = [f(x0)]
x = x0
for _ in range(iterations):
    grad = f_prime(x)
    x = x - eta * grad
    x_vals.append(x)
    y_vals.append(f(x))

# Set up the figure and axis
fig, ax = plt.subplots()
x_range = np.linspace(-5, 5, 400)
y_range = f(x_range)
line, = ax.plot(x_range, y_range, 'b-', label='f(x) = x²')
point, = ax.plot([], [], 'ro', markersize=10)  # moving point
path, = ax.plot([], [], 'r--', alpha=0.5)  # descent path

ax.set_xlim(-5, 5)
ax.set_ylim(0, 25)
ax.set_title("Gradient Descent on f(x) = x²")
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.legend()

# Animation function
def update(frame):
    point.set_data([x_vals[frame]], [y_vals[frame]])
    path.set_data(x_vals[:frame+1], y_vals[:frame+1])
    return point, path

ani = animation.FuncAnimation(fig, update, frames=len(x_vals), interval=300, blit=True)


# Save the animation as a GIF
gif_path = "gradient_descent_x2.gif"
ani.save(gif_path, writer=PillowWriter(fps=2))

gif_path

plt.close(fig)  # Prevent static display
ani
```

```python
import numpy as np
import pandas as pd

# Define the function and its derivative
def f(x):
    return x**4 - 3*x**2 + 2

def f_prime(x):
    return 4*x**3 - 6*x

# Parameters
eta = 0.01         # learning rate
x = 2.5            # starting point
epsilon = 0.7    # stopping threshold for gradient
max_iterations = 1000

# Store values for each step
steps = [(0, x, f(x), f_prime(x))]
iteration = 1

while iteration <= max_iterations:
    grad = f_prime(x)
    steps.append((iteration, x, f(x), grad))
    x = x - eta * grad
    if abs(grad) < epsilon:
        break
    iteration += 1

# Create a DataFrame with formulas in the header
df = pd.DataFrame(steps, columns=[
    "Iteration",
    "x (position)",
    "f(x) = x⁴ - 3x² + 2",
    "f'(x) = 4x³ - 6x"
])

# Display the table
print(df.to_string(index=False))
```

```python
import numpy as np
import pandas as pd

# Define the function and its derivative
def f(x):
    return (x + 2)**2

def f_prime(x):
    return 2 * (x + 2)

# Parameters
eta = 0.2          # learning rate (higher, so it converges faster)
x = 0              # starting point (far from minimum at x=2)
epsilon = 0.001    # stopping condition
max_iterations = 10

# Track steps
steps = [(0, x, f(x), f_prime(x))]
iteration = 1

while iteration <= max_iterations:
    grad = f_prime(x)
    if abs(grad) < epsilon:
        break
    x = x - eta * grad
    steps.append((iteration, x, f(x), grad))
    iteration += 1

# Make DataFrame
df = pd.DataFrame(steps, columns=[
    "Iteration",
    "x (position)",
    "f(x) = (x + 2)²",
    "f'(x) = 2(x + 2)"
])

# Show the table
print(df.to_string(index=False))
```
